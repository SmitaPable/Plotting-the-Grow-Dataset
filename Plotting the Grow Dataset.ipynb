```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

# Read data from CSV file into a DataFrame
df = pd.read_csv("Growlocations.csv")

# Display the original DataFrame
df

# Remove rows with invalid coordinates
df = df[(df['Latitude'] >= -90) & (df['Latitude'] <= 90) & (df['Longitude'] >= -180) & (df['Longitude'] <= 180)]


# Filter the DataFrame based on latitude and longitude ranges
filtered_df = df[
    (df['Latitude'] >= 50.681) & (df['Latitude'] <= 57.985) &
    (df['Longitude'] >= -10.592) & (df['Longitude'] <= 1.6848)
]

# Display the filtered DataFrame
filtered_df

# Open the UK map image
uk_map = Image.open('map7.png')

# Create a figure and axis with a 10 by 10 figsize
fig, ax = plt.subplots(figsize=(10, 10))

# Display the UK map image
ax.imshow(uk_map)

# Set the title of the plot
plt.title('Uk Map')

# Show the plot
plt.show()

# Create a scatter plot of the filtered data
fig, ax = plt.subplots(figsize=(10, 10))
plt.scatter(filtered_df['Longitude'], filtered_df['Latitude'], c='blue', marker='o', alpha=0.7, zorder=1)
plt.title('Scattered Data')

# Create a figure and axis for plotting the filtered data on the UK map
fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(uk_map, extent=[-10.592, 1.6848, 50.681, 57.985])

# Scatter plot of filtered data on the UK map
ax.scatter(filtered_df['Longitude'], filtered_df['Latitude'], color='blue', marker='o', s=10)

# Set equal aspect ratio for the plot
ax.set_aspect('equal')

# Set titles and labels
plt.title('Plotting Grow Data')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Show the plot
plt.show()
```
